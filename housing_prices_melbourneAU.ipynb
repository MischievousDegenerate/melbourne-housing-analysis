# Exploratory Data Analysis (EDA)

```python
"""
Goal: Identifying what conditions influence pricing, hereunder the interrelationship or correlation between variables such as the sqft of living and price.
Examination of the avg house price pr. zipcode as well as other underlying trends and contributors.

"""
```

```python
# Task 1: Counting the number of unique values in the zipcode column.
unique_zipcodes_count = housing_df["zipcode"].nunique()
f"There are {unique_zipcodes_count} different zipcodes in Melbourne Australia"
```

```python
# Task 2: Finding the average house price for each zipcode.
avg_price_per_zipcode = housing_df.groupby("zipcode")["price"].mean()

# Resetting the index
avg_price_per_zipcode = avg_price_per_zipcode.reset_index()

# Rename the columns for clarity
avg_price_per_zipcode.columns = ["zipcode", "average_price"]

# Create descending list of the avg_price_per_zipcode df
avg_price_per_zipcode_descending = avg_price_per_zipcode.sort_values(by="average_price", ascending=False)

# Display descending df
avg_price_per_zipcode_descending
```

```python
# Task 3: Finding the average number of bedrooms for houses with waterfront views (waterfront == 1).

# Filter for houses with waterfront views
waterfront_houses = housing_df[housing_df["waterfront"] == 1]

# Calculate the average number of bedrooms for houses with waterfront views
avg_bedrooms_waterfront = waterfront_houses["bedrooms"].mean()

# Display
f"The average number of bedrooms for houses with waterfront views is: {avg_bedrooms_waterfront}"
```

```python
# Task 4: Determining the percentage of houses with more than 2 floors out of the dataset.

# Convert the "floors" column to numeric (if data is stored as strings)
housing_df["floors"] = housing_df["floors"].astype(int)

# Filter houses with more than 2 floors
houses_more_than_two_floors = housing_df[housing_df["floors"] > 2]

# Calculate the percentage of houses with more than 2 floors
house_percentage_more_than_two_floors = (len(houses_more_than_two_floors) / len(housing_df["floors"])) * 100

# Display
f"Percentage of houses with more than 2 floors: {house_percentage_more_than_two_floors:.2f}%"
```

```python
# Task 5: Checking the correlation between sqft_living and price.
# the corr() function computes the Pearson correlation coefficient: +1 = perfect correlation, -1 = perfect negative correlation, and 0: no correlation
correlation = housing_df["sqft_living"].corr(housing_df["price"])
f"The correlation between 'sqft_living' and 'price' is: {correlation:.2f}. This indicates a strong positive correlation."
```

```python
# To visualize the correlation between sqft of living and price, we can create a scatter plot with a regression line:
import matplotlib.pyplot as plt
import numpy as np

# Scatter plot of "sqft_living" vs "price"
x = housing_df["sqft_living"]
y = housing_df["price"]
plt.scatter(x, y, alpha=0.5, label="Data")

# Adding a linear regression line
slope, intercept = np.polyfit(x, y, 1)
regression_line = slope * x + intercept

# Plot regression line
plt.plot(x, regression_line, color="red", label="Regression Line")

# Titles and labels
plt.title("Scatter Plot of Price vs. Sqft Living")
plt.xlabel("Sqft Living")
plt.ylabel("Price")
plt.legend()

plt.show()
print(f"Slope: {slope:.2f}")
```

```python
# A slope value of 280.62 AUD suggests that if a house was to have 1 more sqft, it'd be expected to cost 280.62 AUD more on average.
```

# Data Manipulation

```python
# Task 6: Creating a new column called house_age that calculates the age of the house based on yr_built.

from datetime import datetime

# Get current year
current_year = datetime.now().year

# Calculate the age of each house
housing_df["house_age"] = current_year - housing_df["yr_built"]

# Display the first few rows to verify
housing_df[["yr_built", "house_age"]].head()
```

```python
housing_df["house_age"].describe()
```

```python
# Task 7: Creating another column called is_renovated that indicates whether a house has been renovated (1 for yes, 0 for no) based on the yr_renovated column.

housing_df["is_renovated"] = housing_df["yr_renovated"].apply(lambda x: 1 if x > 0 else 0)

# Display the first few rows to verify the new column was implemented successfully.
housing_df[["yr_renovated", "is_renovated"]].head()
```

```python
# Count the number of renovated houses
num_renovated = housing_df["is_renovated"].sum()
print(f"The number of renovated houses is: {num_renovated}")
print(f"The percentage of houses that have been renovated based on the dataset is {num_renovated/(len(housing_df))*100:.2f}%")
```

```python
# Task 8: Adding a column called price_per_sqft by dividing price by sqft_living
housing_df["price_per_sqft"] = housing_df["price"] / housing_df["sqft_living"]

# Display new column:
(housing_df[["price_per_sqft", "sqft_living"]].head())
```

```python
housing_df["price_per_sqft"].describe()
```

```python
# Task 9: Filtering the dataframe to only show houses with more than 3 bathrooms and a grade above 8.

filtered_houses = housing_df[(housing_df["bathrooms"] > 3) & (housing_df["grade"] > 8)]

filtered_houses.head()
```

```python
# Count the number of houses that meet the criteria
count_filtered_houses = len(filtered_houses)

print(f"The number of houses with more than 3 bathrooms and a grade above 8 is: {count_filtered_houses}")
print(f"Which accounts for {count_filtered_houses/len(housing_df)*100:.2f}% of the df")
```

```python
# What does the aggregate price of the houses with more than 3 bathrooms and a grade above 8 constitute for the collective house price of the total dataset
agg_filtered_house_price = filtered_houses["price"].sum()
agg_house_price = housing_df["price"].sum()
print(f"The collective filtered house prices accounts for {agg_filtered_house_price/agg_house_price*100:.2f}% of the df ")
```

# Machine Learning Model on prediction of house prices (in-sample testing)
creating a decision tree regressor

```python
housing_df.head()
```

```python
# dropna drops missing values (think of na as "not available")
ml_housing_df1 = housing_df.dropna(axis=0)

# define "price" as our prediction target or y variable
# Note you can also use dot-notation to pull out a single column of values fx y = machine_learning_housing_df.Price
y1 = ml_housing_df1["price"]

# Choosing Features
# The columns that are inputted into our model (and later use) to make predictions) are called "features." In our case, those would be the columns used to determine the home price. Sometimes, you will use all columns except the target as features. Other times you'll be better off with fewer features.

housing_features1 = ["yr_built", "sqft_living", "lat", "long", "grade", "condition", "bedrooms", "bathrooms", "floors", "waterfront"]

# By convention, this data (features) is called X.
X1 = ml_housing_df1[housing_features1]
```

```python
""" 
Building Your Model
You will use the scikit-learn library to create your models. When coding, this library is written as sklearn, as you will see in the sample code. Scikit-learn is easily the most popular library for modeling the types of data typically stored in DataFrames.

The steps to building and using a model are:

Define: What type of model will it be? A decision tree? Some other type of model? Some other parameters of the model type are specified too.
Fit: Capture patterns from provided data. This is the heart of modeling.
Predict: Just what it sounds like
Evaluate: Determine how accurate the model's predictions are.
"""
```

```python
from sklearn.tree import DecisionTreeRegressor

# Define model. Specify a number for random_state to ensure same results each run
melbourne_model_1 = DecisionTreeRegressor(random_state=1)

# Fit model
melbourne_model_1.fit(X1, y1)
```

```python
#ml_housing_df1.info()
```

```python
print("Making predictions for the following 5 houses:")
print(X1.head())
print("The predictions are")
print(melbourne_model_1.predict(X1.head()))
```

```python
y1.head()
```

```python
# Calculating the mean absolute error (MAE) and comparing it to the average house price from the training data

from sklearn.metrics import mean_absolute_error

print("The mean absolute error for in-sample testing")
predicted_home_prices = melbourne_model_1.predict(X1)
mae = mean_absolute_error(y1, predicted_home_prices)
print(mae, end="\n\n")

print("The average house price for in-sample testing (training set)")
avg_training_price = y1.mean()
print(avg_training_price, end="\n\n")

print(f"The machine is on average off by: {mae/avg_training_price * 100:.2f}%")
```

```python
import matplotlib.pyplot as plt

# Scatter plot of "real price" vs "estimated price"
plt.scatter(y1, predicted_home_prices, alpha=0.5)
plt.title("Scatter Plot of real price vs estimated price")
plt.xlabel("Real price")
plt.ylabel("estimated price")
plt.show()
print("The scatter plot is near a perfect linear curve as we trained and tested the model on the same data set.\nIf we had featured all the variables in the training of our model the scatter plot would have been a perfect linear curve and the machine would on average be off by a value of 0%. In other words it would be 100% accurate in predicting house prices, but only on that same training set.")
```

# Machine Learning Model on prediction of house prices (Out-of-sample testing) & validation data

Splitting Training Data and building a decision tree regressor

```python
# dropna drops missing values (think of na as "not available")
ml_housing_df2 = housing_df.dropna(axis=0)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
# Target: "price"
y2 = ml_housing_df2["price"]

# Features:
housing_features2 = ["yr_built", "sqft_living", "lat", "long", "grade", "condition", "bedrooms", "bathrooms", "floors", "waterfront"]
X2 = ml_housing_df2[housing_features2]

# split data into training and validation data, for both features and target. The split is based on a random number generator. Supplying a numeric value to the random_state argument guarantees we get the same split every time we run this script.

train_X, val_X, train_y, val_y = train_test_split(X2, y2, random_state=0)

# Define model
melbourne_model_2 = DecisionTreeRegressor()

# Fit model
melbourne_model_2.fit(train_X, train_y)

# get predicted prices on validation data
val_predictions = melbourne_model_2.predict(val_X)
print(val_predictions)
```

```python
val_y.head()
```

```python
# Calculating the mean absolute error (MAE) and the average house price of the validation data and comparing the results to find the average machine error

print("The mean absolute error for out-of-sample")
MAE = mean_absolute_error(val_y, val_predictions)
print(MAE, end="\n\n")
print("The average house price for the validation data")
avg_validation_price = val_y.mean()
print(avg_validation_price, end="\n\n")
print(f"The machine is on average off by {MAE/avg_validation_price * 100:.2f}%")
```

```python
# Creating a scatter plot for real validation house price and estimated validation house price.

import matplotlib.pyplot as plt

# Scatter plot of "real price" vs "estimated price"
plt.scatter(val_y, val_predictions, alpha=0.5)
plt.title("Scatter Plot of real price vs estimated price")
plt.xlabel("Real price")
plt.ylabel("estimated price")
plt.show()
```
